## 4. Minimal Working Example

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Create an empty graph with no nodes and no edges.
G=nx.Graph()

print(G.nodes())
print(G.edges())

print(type(G.nodes()))
print(type(G.edges()))
```

```python
#Add nodes
G=nx.Graph()
# adding just one node:
G.add_node("a")
# a list of nodes:
G.add_nodes_from(["b","c"])

print("Nodes of graph: ")
print(G.nodes())
print("Edges of graph: ")
print(G.edges())
```

```python
#Add edges
G=nx.Graph()
G.add_node("a")
G.add_nodes_from(["b","c"])

G.add_edge(1,2)
edge = ("d", "e")
G.add_edge(*edge)
edge = ("a", "b")
G.add_edge(*edge)

print("Nodes of graph: ")
print(G.nodes())
print("Edges of graph: ")
print(G.edges())
```

```python
G.add_edges_from([("a","c"),("c","d"), ("a",1), (1,"d"), ("a",2)])
nx.draw(G)
plt.savefig("simple_path.png") # save as png
plt.show() # display
```

```python
G=nx.path_graph(4)

print("Nodes of graph: ")
print(G.nodes())
print("Edges of graph: ")
print(G.edges())
nx.draw(G)
plt.savefig("path_graph1.png")
plt.show()
```

## 5. Example of typical use-cases

### Example 1: Simple Network(Relationship)
This is a simple example of showing the network(relationship) between different fruits.<br />

```python
import networkx as nx
import matplotlib.pyplot as plt

#Create an empty Graph called G
G=nx.Graph()

#add edges to graph G
G.add_edge('Apple', 'Mango') 
G.add_edge('Mango', 'Orange') 
G.add_edge('Grape', 'Banana') 
G.add_edge('Banana', 'Apple') 
G.add_edge('Orange', 'Plum') 
G.add_edge('Pinapple', 'Plum')
G.add_edge('Mango', 'Plum')
G.add_edge('Pinapple', 'Orange')

# add nodes to graph with color=red
G.add_nodes_from(['Apple', 'Banana','Orange','Grape','Mango', 'Pinapple','Plum'])

#We want the graph layouts to be shell shape
graph_shape = nx.shell_layout(G)

#we add blue nodes into graph G. node_color controls for the color of each node, and alpha controls the darkness of that color
nx.draw_networkx_nodes(G, graph_shape, node_size=800, node_color=['red','blue','yellow'], alpha=0.2)
plt.show()
```

```python
#we add network edges into graph G
nx.draw_networkx_edges(G, graph_shape)
plt.show()
```

```python
#we add labels to graph G, and you can adjust the font and its size.
nx.draw_networkx_labels(G, graph_shape, font_size=12, font_family='sans-serif')
plt.show()
```

```python
#run the above 3 stes together and show the network about relationship between fruits.
nx.draw_networkx_nodes(G, graph_shape, node_size=1000, node_color=['red','blue','yellow'], alpha=0.3)
nx.draw_networkx_edges(G, graph_shape)
nx.draw_networkx_labels(G, graph_shape, font_size=10, font_family='sans-serif')

plt.show()
```

### Example 2: Random Geometric Graph
In graph theory, a random geometric graph (RGG) is the mathematically simplest spatial network, namely an undirected graph constructed by randomly placing N nodes in some metric space (according to a specified probability distribution) and connecting two nodes by a link if and only if their distance is in a given range, e.g. smaller than a certain neighborhood radius, r.<br />
A real-world application of RGGs is the modeling of ad hoc networks.<br />

The random geometric graph model places n nodes uniformly at random in the unit cube. Two nodes are joined by an edge if the Euclidean distance between the nodes is at most radius.

With package networkx, you can make random geometric graph using the following function.<br />
random_geometric_graph(n, radius, dim=2, pos=None)<br />

Parameters:	<br />
n (int) â€“ Number of nodes<br />
radius (float) â€“ Distance threshold value<br />
dim (int, optional) â€“ Dimension of graph<br />
pos (dict, optional) â€“ A dictionary keyed by node with node positions as values.<br />


```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.random_geometric_graph(300, 0.13)
# position is stored as node attribute data for random_geometric_graph
pos = nx.get_node_attributes(G, 'pos')

# find node near center (0.3, 0.2)
dmin = 1
ncenter = 0
for n in pos:
    x, y = pos[n]
    d = (x - 0.3)**2 + (y - 0.2)**2
    if d < dmin:
        ncenter = n
        dmin = d

# color by path length from node near center
p = nx.single_source_shortest_path_length(G, ncenter)

plt.figure(figsize = (8, 8))
nx.draw_networkx_edges(G, pos,nodelist = [ncenter], alpha = 0.3)
nx.draw_networkx_nodes(G, pos, nodelist = p.keys(),
                       node_size = 75,
                       node_color = list(p.values()),
                       cmap = plt.cm.PuRd_r)

plt.xlim(-0.05, 1.05)
plt.ylim(-0.05, 1.1)

font = {'fontname'   : 'fantasy',
        'color'      : 'k',
        'fontweight' : 'bold',
        'fontsize'   : 16}
plt.title("Random Geometric Graph", font)

plt.text(0.5, 0.97, "Number of nodes = 300",
         horizontalalignment = 'center',
         transform = plt.gca().transAxes)
plt.text(0.5, 0.94,  "Distance threshold value = 0.13",
         horizontalalignment = 'center',
         transform = plt.gca().transAxes)

    
plt.axis('off')
plt.savefig('random_geometric_graph.png')
plt.show()
```
